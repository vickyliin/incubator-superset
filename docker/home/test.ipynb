```python
from sqlalchemy import create_engine
import pandas as pd
```

```python
engine = create_engine('mysql://superset:superset@mysql:3306/superset')
```

```python
df = pd.read_csv('./gc_activities_kr.csv', index_col='ACTIVITY_ID')
for col in ['UPDATE_DATE', 'CREATION_DATE', 'START_TIME_GMT']:
    df[col] = pd.to_datetime(df[col])
df.head()
```

```python
df[['START_TIME_GMT', 'ACTIVITY_TYPE', 'START_LONGITUDE', 'START_LATITUDE']] \
        .to_sql('gc_activities_kr', engine, chunksize=1000, if_exists='replace')
```
